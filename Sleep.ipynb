```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weekdays = pd.read_csv('Sleep_Weekdays.csv')
weekends = pd.read_csv('Sleep_Weekend.csv')
```

```python
merged_files = pd.merge(weekdays, weekends, on='Year', how='inner')
df = pd.DataFrame(merged_files)
```

```python
graph = df.plot(
    x='Year', 
    y =['Estimate_x', 'Estimate_y'],
    color =['#C72C41', '#1B9CFC'],
    figsize = (8,5)
    )

plt.suptitle("Avg sleeping hours per day in the US (2003-2022)", fontsize=14, fontweight='bold')
plt.title("BLS, American Time Use Survey (ATUS)")

labels = ['Weekdays', 'Weekends']
graph.legend(labels)
graph.set_xticks(df['Year'])
graph.set_xticklabels(df['Year'], rotation=35)
graph.set_xlabel('Year')
graph.set_ylabel('Hours')
graph.grid(linewidth = 1.5, alpha = 0.20)
plt.show()
plt.savefig('my_figure.png')
```
